<h3>Break down the "words". It's not so hard.</h3>

```python
pkmn = pd.read_csv('Pokemon.csv')
```

```python
# ignore - settings
pkmn.columns = pkmn.columns.str.upper().str.replace('_', '') #change into upper case
pkmn = pkmn.set_index('NAME') #ignore
pkmn['TYPE 2'].fillna(pkmn['TYPE 1'], inplace=True) #ignore
```

```python
pkmn[pkmn['LEGENDARY']==True]  #legendary pokemons
```

```python
print(pkmn.loc['Bulbasaur'])
```

```python
plt.scatter(pkmn['HP'],pkmn['ATTACK'])
```

```python
sns.boxplot(data=pkmn);
```

```python
pkmn.corr()
```

```python
sns.heatmap(pkmn.corr())
```

```python
plt.figure(figsize=(10,6)) #manage the size of the plot
sns.heatmap(pkmn.corr(),annot=True) #df.corr() makes a correlation matrix and sns.heatmap is used to show the correlations heatmap
plt.show()
```

```python
a=pkmn.groupby(['GENERATION','TYPE 1']).count().reset_index()
a=a[['GENERATION','TYPE 1','TOTAL']]
a=a.pivot('GENERATION','TYPE 1','TOTAL')
a[['Water','Fire','Grass','Dragon','Normal','Rock','Flying','Electric']].plot(color=['b','r','g','#FFA500','brown','#6666ff','#001012','y'],marker='o')
fig=plt.gcf()
fig.set_size_inches(12,6)
plt.show()
```

```python
pkmn["ATTACK"].hist()
plt.xlabel('Attack') #set the xlabel name
plt.ylabel('Count') #set the ylabel name
```

### Forecast in 2 seconds

```python
peyton = pd.read_csv('townhall_peyton.csv')
peyton.head()
# ds = dates # y = web_traffic
```

```python
m = Prophet()
m.fit(peyton)
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
fig1 = m.plot(forecast)
```

```python
m = Prophet(weekly_seasonality=False)
m.add_seasonality(name='monthly', period=30.5, fourier_order=5)
forecast = m.fit(peyton).predict(future)
fig = m.plot_components(forecast)
```
